### 1. Interactive AI Gemini ChatBot with memory information.

```python
# manage imports

import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_classic.chains import LLMChain
from langchain_classic.memory import ConversationBufferMemory
```

```python
# lets establish the Google Gemini API Key and LLM

%cd .env
load_dotenv()

llm = ChatGoogleGenerativeAI(
    model = 'gemini-2.0-flash-exp',
    google_api_key = os.getenv('GENAI_API_KEY'),
    temperature = 0.7 # this parameter allows being more creative in its responses.
)
```

```python
#config memory
memory = ConversationBufferMemory(memory_key = 'chat_history' , return_messages = True)
```

```python
# creating a chat's template

template = """ the following is a friendly conversation between an Ai assistant and human.
Conversation history:
{chat_history}
Human: {input}
Assistant:"""
```

```python
prompt = PromptTemplate(
    input_variables = ['chat_history' , 'input'],
    template = template
)
```

```python
# this step is the elaboration of the chain with the llm and our memory template

conversation = LLMChain(
    llm = llm,
    prompt = prompt,
    memory = memory,
    verbose = True
)
```

```python
# now we just need to test a fluid conversation with some questions

print(conversation.predict(input = 'Hi, I wanna know who is the fifa world cup champion currently'))
```

In this case I will give a random (Pepito) name as a proove for confirm if it may memorize and include the info requested.

```python
print(conversation.predict(input = 'Mmmm ok, so tell me the top 5 history players for the soccer national team. Just in case my name is Pepito'))
```

```python
print(conversation.predict(input = 'I think its fine but Id like to know hows the tshirt of Argentina?'))
```

```python
print(conversation.predict(input = 'Whats my name?'))
```

### Conclusion

The AI GEMINI BOT was created with success and it is able to remember the information given.

# 2. Sentiment Analysis with the AI GEMINI ChatBot.

```python
from langchain_core.prompts import PromptTemplate
from langchain_classic.chains import LLMChain

#Lets create a template for sentiment analysis

sentiment_template = ''' Please analyze the sentiment of the following text and classify it with positive, neutral or negative.
Also give and explanation of your analysis.

Text: {text}

Format response:

Sentiment: [Classification]

Explanation: [Your analysis]'''
```

```python
sentiment_prompt = PromptTemplate(
    input_variable = ['texto'],
    template = sentiment_template
)
```

```python
# we use the chain and we keep our GEMINI LLM

sentiment_chain = LLMChain( llm = llm , prompt = sentiment_prompt)
```

```python
# analysis example of a sushi reaturant

text_example = '''If you're looking for an unforgettable sushi experience in the heart of the city, Sakura Sushi & Bar is the place to be! From the moment you walk in, the ambiance strikes a perfect balance between modern chic and traditional Japanese aesthetics. The staff were friendly and welcoming, setting the tone for what turned out to be a stellar meal.

We started with the Edamame (perfectly salted and just the right amount of crunch), followed by the Miso Soup, which was rich and comforting. But let’s get to the stars of the show—the sushi.

The Omakase was an absolute highlight. Each piece of fish was impeccably fresh, and you could taste the care and precision that went into crafting each roll. The Toro Nigiri melted in my mouth, and the Spicy Tuna Roll had the perfect kick without overpowering the delicate flavors. For the adventurous eater, the Sakura Special Roll, with its combination of eel, avocado, and a sweet soy glaze, is a must-try!

The restaurant also offers a great selection of cocktails. I had the Yuzu Mule, which paired perfectly with the meal. A great balance of tart and sweet with a little kick from the ginger.

The only minor downside is that it's a bit on the pricier side, but the quality and experience definitely make it worth it. The intimate setting also fills up quickly, so reservations are a good idea.'''
```

```python
# we use the recommended method "predict"

from IPython.display import display

result = sentiment_chain.predict(text = text_example)

print(result)
```

```python
# loading the data
loader = TextLoader('example_data.txt' , encoding = 'utf-8')
documents = loader.load()
print(documents[0])
```

```python
# now we will divide in chunks our document

text_splitter = CharacterTextSplitter(chunk_size = 200 , chunk_overlap = 20) # here we maintain the context
#between chunks over 20 characters
texts = text_splitter.split_documents(documents)
```

```python
# embeddings

embeddings = GoogleGenerativeAIEmbeddings(
    model = 'models/embedding-001',
    google_api_key = os.getenv('GENAI_API_KEY')
)
```

```python
# this step is for vectorial database

db = FAISS.from_documents(texts, embeddings)
```

```python
# lets make a query and test our implementation

query = 'what is soccer?'
docs = db.similarity_search(query)
print(docs[0].page_content)
```

# 4. Creation of an AI GEMINI Agent

Here I will generate an AI Agent with some functions:

```python
# imports
from langchain_classic.agents import Tool, initialize_agent, AgentType
from datetime import datetime
```

```python
# function for getting the current time

def get_current_time(_input = None): # takes the argument but does not use it
    return datetime.now().strftime('%H:%M:%S') 
```

```python
# function for calculating the sum of numbers

def calculate_sum(numbers_str):
    try:
        numbers = [float(n) for n in numbers_str.split()]
        return str(sum(numbers))
    except Exception as error:
        return 'Try again'
        
```

```python
# defining the personalizaed tools

tools = [

    Tool(
        name = 'Current Time',
        func = get_current_time,
        description = 'Allows us to know the time'
    ),

    Tool(
        name = 'Calculator',
        func = calculate_sum,
        description = 'Sum calculator'
    )
    
]
```

```python
# config agent

agent = initialize_agent(
    tools = tools,
    llm = llm,
    agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION, # its the agent config for default
    verbose = True
)
```

```python
# testing
print(agent.run('What time is it?'))
```

```python
print(agent.run('sum the numbers 10 20 30 40 50'))
```

Will change the agent response adding commas between numbers?

```python
print(agent.run('sum 10, 20, 30, 40, 50'))
```

Finally we can confirm that the result does not change even there is commas or not and more import the agent can identify the calculate_sum function and realize that must use it.
